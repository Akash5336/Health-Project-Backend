```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import matplotlib.pyplot as plt 
%matplotlib inline
import csv
```

```python
df = pd.read_csv('new_set.csv')
```

```python
df
```

```python
df.replace(float('nan'), np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
disease_list = list(df['disease'].unique())
```

```python
len(disease_list)
```

```python
symptom_list = list(df['symptom'].unique())
len(symptom_list)
```

```python
from collections import defaultdict
disease_sym = defaultdict(list)

k = 0
for i in df['disease'].values:
    disease_sym[i].append(df.iloc[k]['symptom'])
    k += 1
```

```python
disease_sym
```

```python
df.dtypes
```

```python
label_encoder = LabelEncoder()
encoded_values2 = label_encoder.fit_transform(df['symptom'])
print(encoded_values2)
len(encoded_values2)
```

```python
encoded_values = label_encoder.fit_transform(symptom_list)
print(encoded_values)
len(encoded_values)
```

```python
encoded_df = pd.DataFrame(columns = symptom_list)
encoded_df.head()

encoded_df.loc[0] = encoded_values
```

```python
encoded_df
```

```python
onehot_encoder = OneHotEncoder(sparse= False)
encoded_values = encoded_values.reshape(len(encoded_values), 1)
onehot_encoded = onehot_encoder.fit_transform(encoded_values) 

print(onehot_encoded)
```
